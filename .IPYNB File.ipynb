```python

# STEP 1: INSTALL AND IMPORT LIBRARIES

!pip install -q xgboost

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, KFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.feature_selection import SelectKBest, f_regression
from scipy import stats
import warnings
import time
import joblib
import os
warnings.filterwarnings('ignore')

print("="*80)
print("LIBRARIES LOADED SUCCESSFULLY!")
print("="*80)
# STEP 2: UPLOAD DATASET

print("\n" + "="*80)
print("PLEASE UPLOAD YOUR DATASET")
print("="*80)
print("Click on the 'Files' icon on the left sidebar")
print("Click 'Upload' and select: smart_electricity_consumption_dataset.csv")
print("\nAfter uploading, press Enter to continue...")

from google.colab import files
uploaded = files.upload()

if len(uploaded) == 0:
    print("\nNo file uploaded! Please upload the dataset.")
    raise SystemExit("Dataset upload required!")

filename = list(uploaded.keys())[0]
print(f"\nFile uploaded successfully: {filename}")

# STEP 3: LOAD DATASET

df = pd.read_csv(filename)

print("\n" + "="*80)
print("DATASET LOADED SUCCESSFULLY!")
print("="*80)
print(f"Dataset Shape: {df.shape[0]} records, {df.shape[1]} features")
print("\nFirst 5 rows:")
display(df.head())

print("\nColumn Information:")
print(df.info())

print("\nStatistical Summary:")
display(df.describe())

# STEP 4: DATA CLEANING - HANDLE MISSING VALUES

print("\n" + "="*80)
print("DATA CLEANING - HANDLING MISSING VALUES")
print("="*80)

missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

print("Missing Values Analysis:")
missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage
})
display(missing_df[missing_df['Missing Values'] > 0])

df_cleaned = df.copy()

# Fill missing values with median for numeric columns
numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df_cleaned[col].isnull().sum() > 0:
        median_val = df_cleaned[col].median()
        df_cleaned[col].fillna(median_val, inplace=True)
        print(f"Filled missing in '{col}' with median: {median_val:.2f}")

duplicates = df_cleaned.duplicated().sum()
print(f"\nDuplicate Records: {duplicates}")
if duplicates > 0:
    df_cleaned = df_cleaned.drop_duplicates()
    print(f"Removed {duplicates} duplicate records")

remaining_missing = df_cleaned.isnull().sum().sum()
print(f"\nRemaining missing values: {remaining_missing}")
print(f"Final dataset shape: {df_cleaned.shape}")

# STEP 5: FEATURE ENCODING - CRITICAL FIX

print("\n" + "="*80)
print("FEATURE ENCODING - CONVERTING STRINGS TO NUMBERS")
print("="*80)

# ENSURE ALL STRING COLUMNS ARE ENCODED
label_encoders = {}
categorical_cols = ['House_Type', 'Day_of_Week', 'Season']

df_encoded = df_cleaned.copy()

for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    label_encoders[col] = le
    print(f"{col}:")
    for i, category in enumerate(le.classes_):
        print(f"  {category} -> {i}")

print("\nEncoded dataset preview:")
display(df_encoded.head())

# VERIFY: No string columns remain
print("\n" + "="*50)
print("VERIFYING DATA TYPES - ALL SHOULD BE NUMERIC")
print("="*50)
print(df_encoded.dtypes)
print("\nNumber of object/string columns:", sum(df_encoded.dtypes == 'object'))

# If any string columns remain, encode them
if sum(df_encoded.dtypes == 'object') > 0:
    print("\nWARNING: Some string columns still exist! Encoding them...")
    for col in df_encoded.select_dtypes(include=['object']).columns:
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
        label_encoders[col] = le
        print(f"  Encoded {col}")

print("\nFinal data types after encoding:")
print(df_encoded.dtypes.value_counts())

# STEP 6: FEATURE SCALING

print("\n" + "="*80)
print("FEATURE SCALING")
print("="*80)

numeric_features = ['Family_Members_or_Staff', 'Number_of_Rooms_or_Units', 'AC_Usage_Hours',
                    'Fan_Usage_Hours', 'Refrigerator_Usage_Hours', 'Washing_Machine_Usage_Hours',
                    'Water_Motor_Usage_Hours', 'Lighting_Hours', 'Daily_Appliance_Count',
                    'Machinery_Load_kW', 'Machinery_Usage_Hours', 'Outdoor_Temperature_C',
                    'Electricity_Unit_Price']

scaler = StandardScaler()
df_scaled = df_encoded.copy()
df_scaled[numeric_features] = scaler.fit_transform(df_encoded[numeric_features])

print(f"Scaled {len(numeric_features)} numeric features")
print("\nAfter scaling (first 5 rows):")
display(df_scaled[numeric_features].head())

# STEP 7: EXPLORATORY DATA ANALYSIS - EXTENSIVE CHARTS

print("\n" + "="*80)
print("EXPLORATORY DATA ANALYSIS - EXTENSIVE CHARTS")
print("="*80)

# CHART 1: Target Variable Distribution (6 subplots)
print("\n1. Target Variable Distribution - Multiple Views")
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

axes[0, 0].hist(df_cleaned['Daily_Electricity_Consumption_kWh'], bins=50, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Histogram', fontsize=12)
axes[0, 0].set_xlabel('Consumption (kWh)')
axes[0, 0].set_ylabel('Frequency')

sns.kdeplot(df_cleaned['Daily_Electricity_Consumption_kWh'], ax=axes[0, 1], color='darkblue', fill=True)
axes[0, 1].set_title('Kernel Density Plot', fontsize=12)
axes[0, 1].set_xlabel('Consumption (kWh)')

sns.boxplot(x=df_cleaned['Daily_Electricity_Consumption_kWh'], ax=axes[0, 2], color='lightcoral')
axes[0, 2].set_title('Box Plot', fontsize=12)
axes[0, 2].set_xlabel('Consumption (kWh)')

stats.probplot(df_cleaned['Daily_Electricity_Consumption_kWh'], dist="norm", plot=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot', fontsize=12)

sns.violinplot(x=df_cleaned['Daily_Electricity_Consumption_kWh'], ax=axes[1, 1], color='lightgreen')
axes[1, 1].set_title('Violin Plot', fontsize=12)
axes[1, 1].set_xlabel('Consumption (kWh)')

sns.ecdfplot(df_cleaned['Daily_Electricity_Consumption_kWh'], ax=axes[1, 2], color='purple')
axes[1, 2].set_title('Cumulative Distribution', fontsize=12)
axes[1, 2].set_xlabel('Consumption (kWh)')
axes[1, 2].set_ylabel('Cumulative Probability')

plt.suptitle('Target Variable Analysis - Multiple Views', fontsize=16)
plt.tight_layout()
plt.show()

# CHART 2: Appliance Usage Distributions
print("\n2. Appliance Usage Distributions")
appliance_cols = ['AC_Usage_Hours', 'Fan_Usage_Hours', 'Refrigerator_Usage_Hours',
                  'Washing_Machine_Usage_Hours', 'Water_Motor_Usage_Hours', 'Lighting_Hours']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
for idx, col in enumerate(appliance_cols):
    row, col_idx = idx // 3, idx % 3
    sns.histplot(df_cleaned[col], bins=30, kde=True, ax=axes[row, col_idx], color='teal')
    axes[row, col_idx].set_title(f'Distribution of {col}')
    axes[row, col_idx].set_xlabel('Hours')
    axes[row, col_idx].set_ylabel('Frequency')
fig.suptitle('Appliance Usage Distributions', fontsize=16)
plt.tight_layout()
plt.show()

# CHART 3: Pie Chart - House Type Distribution
print("\n3. Pie Chart - House Type Distribution")
house_counts = df['House_Type'].value_counts()
plt.figure(figsize=(10, 8))
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#ff99cc','#99ccff','#ffb3e6']
plt.pie(house_counts.values, labels=house_counts.index, autopct='%1.1f%%',
        colors=colors[:len(house_counts)], startangle=90, explode=[0.02]*len(house_counts))
plt.title('Distribution of House Types', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()

# CHART 4: Pie Chart - Seasonal Distribution
print("\n4. Pie Chart - Seasonal Distribution")
season_counts = df['Season'].value_counts()
plt.figure(figsize=(10, 8))
colors = ['lightblue', 'lightgreen', 'orange', 'gray']
plt.pie(season_counts.values, labels=season_counts.index, autopct='%1.1f%%',
        colors=colors, startangle=90, explode=[0.02]*len(season_counts))
plt.title('Distribution of Seasons', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()

# CHART 5: Pie Chart - Holiday vs Non-Holiday
print("\n5. Pie Chart - Holiday vs Non-Holiday")
holiday_counts = df['Is_Holiday'].value_counts()
holiday_labels = ['Non-Holiday' if x==0 else 'Holiday' for x in holiday_counts.index]
plt.figure(figsize=(8, 8))
plt.pie(holiday_counts.values, labels=holiday_labels, autopct='%1.1f%%',
        colors=['#66b3ff', '#ff9999'], startangle=90, explode=[0.02, 0.02])
plt.title('Holiday vs Non-Holiday Distribution', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()

# CHART 6: Box Plots for All Numeric Features
print("\n6. Box Plots for All Numeric Features")
numeric_cols_all = df_cleaned.select_dtypes(include=[np.number]).columns.tolist()
if 'Daily_Electricity_Consumption_kWh' in numeric_cols_all:
    numeric_cols_all.remove('Daily_Electricity_Consumption_kWh')

fig, axes = plt.subplots(4, 4, figsize=(20, 16))
axes_flat = axes.flatten()
for idx, col in enumerate(numeric_cols_all[:16]):
    if idx < len(axes_flat):
        sns.boxplot(y=df_cleaned[col], ax=axes_flat[idx], color='skyblue')
        axes_flat[idx].set_title(col, fontsize=10)
        axes_flat[idx].set_ylabel('')
for idx in range(len(numeric_cols_all[:16]), len(axes_flat)):
    axes_flat[idx].axis('off')
plt.suptitle('Box Plots for All Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()

# CHART 7: Correlation Heatmap
print("\n7. Correlation Heatmap")
plt.figure(figsize=(16, 14))
correlation_matrix = df_encoded.select_dtypes(include=[np.number]).corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            fmt='.2f', linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.show()

# CHART 8: Top Correlations Bar Chart
print("\n8. Top Correlations with Target")
target_corr = correlation_matrix['Daily_Electricity_Consumption_kWh'].sort_values(ascending=False)
top_corr = target_corr[1:16]

plt.figure(figsize=(12, 8))
colors = ['green' if x > 0 else 'red' for x in top_corr.values]
sns.barplot(x=top_corr.values, y=top_corr.index, palette=colors)
plt.title('Top 15 Features Correlated with Daily Electricity Consumption', fontsize=14)
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
plt.tight_layout()
plt.show()

# CHART 9: Scatter Plots - Key Features vs Target
print("\n9. Scatter Plots - Key Features vs Target")
temp_df = df_cleaned.copy()
temp_df['Total_Appliance_Hours'] = (df_cleaned['AC_Usage_Hours'] + df_cleaned['Fan_Usage_Hours'] +
                                    df_cleaned['Refrigerator_Usage_Hours'] + df_cleaned['Washing_Machine_Usage_Hours'] +
                                    df_cleaned['Water_Motor_Usage_Hours'] + df_cleaned['Lighting_Hours'])

key_features = ['AC_Usage_Hours', 'Fan_Usage_Hours', 'Outdoor_Temperature_C',
                'Lighting_Hours', 'Daily_Appliance_Count', 'Total_Appliance_Hours']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
for idx, feature in enumerate(key_features):
    row, col_idx = idx // 3, idx % 3
    axes[row, col_idx].scatter(temp_df[feature], temp_df['Daily_Electricity_Consumption_kWh'],
                              alpha=0.4, s=20, c='blue')
    axes[row, col_idx].set_xlabel(feature)
    axes[row, col_idx].set_ylabel('Electricity Consumption (kWh)')
    axes[row, col_idx].set_title(f'{feature} vs Consumption')
    z = np.polyfit(temp_df[feature], temp_df['Daily_Electricity_Consumption_kWh'], 1)
    p = np.poly1d(z)
    axes[row, col_idx].plot(temp_df[feature].sort_values(),
                           p(temp_df[feature].sort_values()), "r--", alpha=0.8)
fig.suptitle('Scatter Plots: Features vs Electricity Consumption', fontsize=16)
plt.tight_layout()
plt.show()

# CHART 10: Pair Plot
print("\n10. Pair Plot - Top Features")
top_5_features = ['AC_Usage_Hours', 'Fan_Usage_Hours', 'Outdoor_Temperature_C',
                  'Daily_Appliance_Count', 'Daily_Electricity_Consumption_kWh']
sns.pairplot(df_cleaned[top_5_features], diag_kind='kde', corner=True)
plt.suptitle('Pair Plot - Top Features', fontsize=16, y=1.02)
plt.show()

# CHART 11: Categorical Analysis
print("\n11. Categorical Analysis")
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

sns.boxplot(x='House_Type', y='Daily_Electricity_Consumption_kWh', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Consumption by House Type', fontsize=12)
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=45, ha='right')

sns.countplot(x='House_Type', data=df, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Count by House Type', fontsize=12)
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=45, ha='right')

sns.boxplot(x='Season', y='Daily_Electricity_Consumption_kWh', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Consumption by Season', fontsize=12)

sns.boxplot(x='Is_Holiday', y='Daily_Electricity_Consumption_kWh', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Consumption by Holiday Status', fontsize=12)
axes[1, 1].set_xticklabels(['Non-Holiday', 'Holiday'])

plt.suptitle('Categorical Variable Analysis', fontsize=16)
plt.tight_layout()
plt.show()

# CHART 12: Seasonal and Holiday Averages
print("\n12. Seasonal and Holiday Averages")
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

seasonal_avg = df.groupby('Season')['Daily_Electricity_Consumption_kWh'].mean()
seasonal_avg.plot(kind='bar', ax=axes[0], color=['blue', 'green', 'red', 'orange'])
axes[0].set_title('Average Consumption by Season', fontsize=12)
axes[0].set_xlabel('Season')
axes[0].set_ylabel('Average Consumption (kWh)')

holiday_avg = df.groupby('Is_Holiday')['Daily_Electricity_Consumption_kWh'].mean()
holiday_avg.index = ['Non-Holiday', 'Holiday']
holiday_avg.plot(kind='bar', ax=axes[1], color=['blue', 'red'])
axes[1].set_title('Average Consumption: Holiday vs Non-Holiday', fontsize=12)
axes[1].set_xlabel('Day Type')
axes[1].set_ylabel('Average Consumption (kWh)')

plt.tight_layout()
plt.show()

# CHART 13: AC Usage vs Temperature Scatter
print("\n13. AC Usage vs Temperature Analysis")
plt.figure(figsize=(12, 6))
scatter = plt.scatter(df_cleaned['Outdoor_Temperature_C'], df_cleaned['AC_Usage_Hours'],
                     c=df_cleaned['Daily_Electricity_Consumption_kWh'],
                     cmap='viridis', alpha=0.6, s=30)
plt.colorbar(scatter, label='Electricity Consumption (kWh)')
plt.xlabel('Outdoor Temperature (°C)')
plt.ylabel('AC Usage Hours')
plt.title('AC Usage vs Temperature (Colored by Consumption)', fontsize=14)
plt.tight_layout()
plt.show()

# CHART 14: Donut Chart - Average Consumption by House Type
print("\n14. Donut Chart - Average Consumption by House Type")
avg_by_house = df.groupby('House_Type')['Daily_Electricity_Consumption_kWh'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 8))
colors = plt.cm.Set3(np.linspace(0, 1, len(avg_by_house)))
wedges, texts, autotexts = plt.pie(avg_by_house.values, labels=avg_by_house.index,
                                   autopct='%1.1f%%', colors=colors, startangle=90,
                                   wedgeprops={'width': 0.7})
plt.title('Average Consumption Distribution by House Type', fontsize=14)
plt.tight_layout()
plt.show()

# STEP 8: FEATURE ENGINEERING

print("\n" + "="*80)
print("FEATURE ENGINEERING")
print("="*80)

df_featured = df_encoded.copy()  # Use encoded dataframe

# 1. Total Appliance Usage Hours
df_featured['Total_Appliance_Hours'] = (df_featured['AC_Usage_Hours'] +
                                        df_featured['Fan_Usage_Hours'] +
                                        df_featured['Refrigerator_Usage_Hours'] +
                                        df_featured['Washing_Machine_Usage_Hours'] +
                                        df_featured['Water_Motor_Usage_Hours'] +
                                        df_featured['Lighting_Hours'])

# 2. Average Appliance Usage per Person
df_featured['Appliance_Hours_per_Person'] = df_featured['Total_Appliance_Hours'] / (df_featured['Family_Members_or_Staff'] + 1)

# 3. Rooms per Person
df_featured['Rooms_per_Person'] = df_featured['Number_of_Rooms_or_Units'] / (df_featured['Family_Members_or_Staff'] + 1)

# 4. High AC Usage Indicator
df_featured['High_AC_Usage'] = (df_featured['AC_Usage_Hours'] > 8).astype(int)

# 5. Temperature Extremes
df_featured['High_Temperature'] = (df_featured['Outdoor_Temperature_C'] > 35).astype(int)
df_featured['Low_Temperature'] = (df_featured['Outdoor_Temperature_C'] < 15).astype(int)

# 6. Appliance Count per Person
df_featured['Appliance_Count_per_Person'] = df_featured['Daily_Appliance_Count'] / (df_featured['Family_Members_or_Staff'] + 1)

# 7. Is Weekend
df_featured['Is_Weekend'] = df_featured['Day_of_Week'].isin([5, 6]).astype(int)

# 8. Efficiency Score
df_featured['Efficiency_Score'] = df_featured['Total_Appliance_Hours'] / (df_featured['Daily_Electricity_Consumption_kWh'] + 1)

# 9. AC to Fan Ratio
df_featured['AC_Fan_Ratio'] = df_featured['AC_Usage_Hours'] / (df_featured['Fan_Usage_Hours'] + 1)

print("New features created:")
new_features = ['Total_Appliance_Hours', 'Appliance_Hours_per_Person', 'Rooms_per_Person',
                'High_AC_Usage', 'High_Temperature', 'Low_Temperature',
                'Appliance_Count_per_Person', 'Is_Weekend', 'Efficiency_Score', 'AC_Fan_Ratio']
for feat in new_features:
    print(f"  - {feat}")

print(f"\nDataset shape after feature engineering: {df_featured.shape}")
display(df_featured[new_features].head())

# VERIFY: All columns are numeric now
print("\nVerifying all columns are numeric:")
print(df_featured.dtypes.value_counts())
if sum(df_featured.dtypes == 'object') > 0:
    print("WARNING: Object columns still exist!")
    print(df_featured.dtypes[df_featured.dtypes == 'object'])


# STEP 9: FEATURE SELECTION - FIXED

print("\n" + "="*80)
print("FEATURE SELECTION")
print("="*80)

# Define all features - ALL NUMERIC NOW
feature_cols = ['House_Type', 'Family_Members_or_Staff', 'Number_of_Rooms_or_Units',
                'AC_Usage_Hours', 'Fan_Usage_Hours', 'Refrigerator_Usage_Hours',
                'Washing_Machine_Usage_Hours', 'Water_Motor_Usage_Hours', 'Lighting_Hours',
                'Daily_Appliance_Count', 'Machinery_Load_kW', 'Machinery_Usage_Hours',
                'Outdoor_Temperature_C', 'Electricity_Unit_Price', 'Day_of_Week', 'Season',
                'Is_Holiday', 'Total_Appliance_Hours', 'Appliance_Hours_per_Person',
                'Rooms_per_Person', 'High_AC_Usage', 'High_Temperature', 'Low_Temperature',
                'Appliance_Count_per_Person', 'Is_Weekend', 'Efficiency_Score', 'AC_Fan_Ratio']

X = df_featured[feature_cols]
y = df_featured['Daily_Electricity_Consumption_kWh']

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print("\nFeature data types:")
print(X.dtypes.value_counts())

# VERIFY: No object/string columns
if sum(X.dtypes == 'object') > 0:
    print("\nERROR: Some columns are still strings!")
    print(X.dtypes[X.dtypes == 'object'])
    raise ValueError("Please encode all string columns first!")

print("\nAll features are numeric! Proceeding with feature selection...")

# Feature selection using SelectKBest
selector = SelectKBest(score_func=f_regression, k=20)
X_selected = selector.fit_transform(X, y)

# Get feature scores
feature_scores = pd.DataFrame({
    'Feature': feature_cols,
    'Score': selector.scores_
}).sort_values('Score', ascending=False)

print("Feature Selection Scores (Top 20):")
display(feature_scores.head(20))

# Select top features
top_features = feature_scores['Feature'].head(15).tolist()
print(f"\nSelected top 15 features for modeling:")
for i, feat in enumerate(top_features, 1):
    score_val = feature_scores[feature_scores['Feature'] == feat]['Score'].values[0]
    print(f"  {i:2}. {feat:35} | Score: {score_val:.2f}")

# CHART 15: Feature Selection Scores
plt.figure(figsize=(12, 8))
sns.barplot(x='Score', y='Feature', data=feature_scores.head(20), palette='viridis')
plt.title('Feature Selection Scores - Top 20 Features', fontsize=14)
plt.xlabel('F-Score')
plt.ylabel('Features')
plt.tight_layout()
plt.show()

# STEP 10: PREPARE DATA FOR MODELING

print("\n" + "="*80)
print("DATA PREPARATION FOR MODELING")
print("="*80)

X_final = df_featured[top_features]
y_final = df_featured['Daily_Electricity_Consumption_kWh']

scaler_final = StandardScaler()
X_scaled = scaler_final.fit_transform(X_final)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_final, test_size=0.2, random_state=42)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Testing set: {X_test.shape[0]} samples")
print(f"Number of features: {X_train.shape[1]}")

print("\nTarget Statistics:")
print(f"Training - Mean: {y_train.mean():.2f}, Std: {y_train.std():.2f}")
print(f"Testing  - Mean: {y_test.mean():.2f}, Std: {y_test.std():.2f}")

# STEP 11: MODEL DEVELOPMENT AND COMPARISON

print("\n" + "="*80)
print("MODEL DEVELOPMENT AND COMPARISON")
print("="*80)

models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100, random_state=42, n_jobs=-1),
    'SVR': SVR()
}

results = {}
predictions = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    start_time = time.time()
    model.fit(X_train, y_train)
    training_time = time.time() - start_time

    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    results[name] = {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2': r2, 'Time': training_time}
    predictions[name] = y_pred

    print(f"  MAE: {mae:.4f}, RMSE: {rmse:.4f}, R2: {r2:.4f}, Time: {training_time:.2f}s")

results_df = pd.DataFrame(results).T
print("\n" + "="*80)
print("MODEL PERFORMANCE COMPARISON")
print("="*80)
display(results_df)

best_model_name = results_df['R2'].idxmax()
best_model = models[best_model_name]
print(f"\nBest Model: {best_model_name}")
print(f"  R2 Score: {results_df.loc[best_model_name, 'R2']:.4f}")
print(f"  RMSE: {results_df.loc[best_model_name, 'RMSE']:.4f}")

# STEP 12: MODEL PERFORMANCE VISUALIZATION

print("\n" + "="*80)
print("MODEL PERFORMANCE VISUALIZATION")
print("="*80)

# CHART 16: Model Comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

results_df['MAE'].sort_values().plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Mean Absolute Error by Model', fontsize=12)
axes[0].set_ylabel('MAE')
axes[0].set_xlabel('Model')
axes[0].tick_params(axis='x', rotation=45)

results_df['RMSE'].sort_values().plot(kind='bar', ax=axes[1], color='lightcoral')
axes[1].set_title('Root Mean Squared Error by Model', fontsize=12)
axes[1].set_ylabel('RMSE')
axes[1].set_xlabel('Model')
axes[1].tick_params(axis='x', rotation=45)

results_df['R2'].sort_values().plot(kind='bar', ax=axes[2], color='lightgreen')
axes[2].set_title('R2 Score by Model', fontsize=12)
axes[2].set_ylabel('R2 Score')
axes[2].set_xlabel('Model')
axes[2].set_ylim(0, 1)
axes[2].tick_params(axis='x', rotation=45)

plt.suptitle('Model Performance Comparison', fontsize=16)
plt.tight_layout()
plt.show()

# CHART 17: Actual vs Predicted
y_pred_best = predictions[best_model_name]

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

axes[0, 0].scatter(y_test, y_pred_best, alpha=0.5, s=20)
axes[0, 0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
axes[0, 0].set_xlabel('Actual Values')
axes[0, 0].set_ylabel('Predicted Values')
axes[0, 0].set_title(f'Actual vs Predicted - {best_model_name}')

residuals = y_test - y_pred_best
axes[0, 1].scatter(y_pred_best, residuals, alpha=0.5, s=20)
axes[0, 1].axhline(y=0, color='r', linestyle='--')
axes[0, 1].set_xlabel('Predicted Values')
axes[0, 1].set_ylabel('Residuals')
axes[0, 1].set_title('Residual Plot')

sns.histplot(residuals, kde=True, ax=axes[1, 0], color='purple')
axes[1, 0].set_xlabel('Residuals')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].set_title('Distribution of Residuals')

stats.probplot(residuals, dist="norm", plot=axes[1, 1])
axes[1, 1].set_title('Q-Q Plot of Residuals')

plt.suptitle(f'Best Model: {best_model_name} - Detailed Analysis', fontsize=16)
plt.tight_layout()
plt.show()

# STEP 13: FEATURE IMPORTANCE ANALYSIS

print("\n" + "="*80)
print("FEATURE IMPORTANCE ANALYSIS")
print("="*80)

if hasattr(best_model, 'feature_importances_'):
    importance = best_model.feature_importances_
    feature_importance_df = pd.DataFrame({
        'Feature': top_features,
        'Importance': importance
    }).sort_values('Importance', ascending=False)

    # CHART 18: Feature Importance
    plt.figure(figsize=(12, 10))
    sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(15), palette='viridis')
    plt.title('Feature Importance Analysis - Top 15 Features', fontsize=14)
    plt.xlabel('Importance')
    plt.ylabel('Features')
    plt.tight_layout()
    plt.show()

    print("Top 10 Most Important Features:")
    display(feature_importance_df.head(10))

    # CHART 19: Cumulative Importance
    feature_importance_df['Cumulative'] = feature_importance_df['Importance'].cumsum()
    plt.figure(figsize=(12, 6))
    plt.plot(range(1, len(feature_importance_df) + 1), feature_importance_df['Cumulative'], 'b-o')
    plt.axhline(y=0.95, color='r', linestyle='--', label='95% Threshold')
    plt.xlabel('Number of Features')
    plt.ylabel('Cumulative Importance')
    plt.title('Cumulative Feature Importance')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# STEP 14: CROSS-VALIDATION

print("\n" + "="*80)
print("CROSS-VALIDATION ANALYSIS")
print("="*80)

kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(best_model, X_scaled, y_final, cv=kf, scoring='r2')

print(f"{best_model_name} - Cross-Validation Results:")
print("-"*50)
for i, score in enumerate(cv_scores, 1):
    print(f"  Fold {i}: R2 = {score:.4f}")
print(f"\nMean CV R2 Score: {cv_scores.mean():.4f}")
print(f"Std CV R2 Score: {cv_scores.std():.4f}")

# CHART 20: Cross-Validation Scores
plt.figure(figsize=(8, 5))
sns.boxplot(x=cv_scores, color='skyblue')
plt.title(f'Cross-Validation Scores - {best_model_name}', fontsize=12)
plt.xlabel('R2 Score')
plt.show()

# STEP 15: HYPERPARAMETER OPTIMIZATION

print("\n" + "="*80)
print("HYPERPARAMETER OPTIMIZATION")
print("="*80)

param_grids = {
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [10, 20, 30, None],
        'min_samples_split': [2, 5, 10]
    },
    'XGBoost': {
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 6, 9],
        'learning_rate': [0.01, 0.1, 0.3]
    },
    'Gradient Boosting': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 0.3],
        'max_depth': [3, 5, 7]
    }
}

if best_model_name in param_grids:
    print(f"Performing Grid Search for {best_model_name}...")

    grid_search = GridSearchCV(
        models[best_model_name],
        param_grids[best_model_name],
        cv=3,
        scoring='r2',
        n_jobs=-1,
        verbose=0
    )
    grid_search.fit(X_train, y_train)

    print(f"\nBest Parameters: {grid_search.best_params_}")
    print(f"Best CV Score: {grid_search.best_score_:.4f}")

    best_model_optimized = grid_search.best_estimator_
    y_pred_opt = best_model_optimized.predict(X_test)

    r2_opt = r2_score(y_test, y_pred_opt)
    rmse_opt = np.sqrt(mean_squared_error(y_test, y_pred_opt))
    mae_opt = mean_absolute_error(y_test, y_pred_opt)

    print(f"\nOptimized Model Performance:")
    print(f"  R2 Score: {r2_opt:.4f} (improvement: {(r2_opt - results[best_model_name]['R2']):.4f})")
    print(f"  RMSE: {rmse_opt:.4f} (improvement: {(results[best_model_name]['RMSE'] - rmse_opt):.4f})")
    print(f"  MAE: {mae_opt:.4f} (improvement: {(results[best_model_name]['MAE'] - mae_opt):.4f})")

    best_model = best_model_optimized
    y_pred_best = y_pred_opt

# STEP 16: SAVE MODEL AND PREPROCESSORS

print("\n" + "="*80)
print("SAVING MODEL AND PREPROCESSORS")
print("="*80)

if not os.path.exists('models'):
    os.makedirs('models')

joblib.dump(best_model, 'models/electricity_consumption_model.pkl')
joblib.dump(scaler_final, 'models/scaler.pkl')
joblib.dump(top_features, 'models/feature_columns.pkl')
joblib.dump(label_encoders, 'models/label_encoders.pkl')

print("Files saved successfully:")
print("  - models/electricity_consumption_model.pkl")
print("  - models/scaler.pkl")
print("  - models/feature_columns.pkl")
print("  - models/label_encoders.pkl")

# STEP 17: PREDICTION FUNCTIONS

print("\n" + "="*80)
print("PREDICTION SYSTEM")
print("="*80)

def predict_consumption(input_data):
    feature_vector = []
    for col in top_features:
        if col in input_data:
            feature_vector.append(input_data[col])
        else:
            feature_vector.append(0)

    features_scaled = scaler_final.transform([feature_vector])
    prediction = best_model.predict(features_scaled)[0]
    return prediction

def generate_recommendations(input_data, prediction, unit_price=35):
    recommendations = []

    ac_hours = input_data.get('AC_Usage_Hours', 0)
    if ac_hours > 8:
        recommendations.append({
            'category': 'Air Conditioner',
            'recommendation': 'Reduce AC usage to 6-8 hours/day. Set temperature to 26°C and use timer.',
            'potential_savings': '5-10% of daily consumption'
        })

    fan_hours = input_data.get('Fan_Usage_Hours', 0)
    if fan_hours > 12:
        recommendations.append({
            'category': 'Fans',
            'recommendation': 'Reduce fan usage. Use ceiling fans instead of standing fans.',
            'potential_savings': '2-5% of daily consumption'
        })

    wash_hours = input_data.get('Washing_Machine_Usage_Hours', 0)
    if wash_hours > 1.5:
        recommendations.append({
            'category': 'Washing Machine',
            'recommendation': 'Use cold water wash cycle. Wait for full load before running.',
            'potential_savings': '2-3% of daily consumption'
        })

    light_hours = input_data.get('Lighting_Hours', 0)
    if light_hours > 10:
        recommendations.append({
            'category': 'Lighting',
            'recommendation': 'Switch to LED bulbs. Use natural lighting during daytime.',
            'potential_savings': '1-3% of daily consumption'
        })

    temp = input_data.get('Outdoor_Temperature_C', 0)
    if temp > 35:
        recommendations.append({
            'category': 'Temperature Management',
            'recommendation': 'High temperature detected. Pre-cool early morning. Use curtains to block heat.',
            'potential_savings': '3-8% of daily consumption'
        })

    recommendations.append({
        'category': 'General Tips',
        'recommendation': 'Monitor peak usage hours (6-10 PM). Shift heavy loads to off-peak times.',
        'potential_savings': '5-10% of daily consumption'
    })

    monthly_consumption = prediction * 30
    monthly_cost = monthly_consumption * unit_price

    recommendations.append({
        'category': 'Cost Analysis',
        'recommendation': f'Current monthly cost: Rs. {monthly_cost:,.2f}. '
                         f'Saving 10% could reduce bill by Rs. {monthly_cost * 0.1:,.2f}/month.',
        'potential_savings': '10-15% of monthly bill'
    })

    return recommendations

# Example prediction
print("\n" + "="*80)
print("EXAMPLE PREDICTION")
print("="*80)

example_input = {
    'House_Type': 0,
    'Family_Members_or_Staff': 5,
    'Number_of_Rooms_or_Units': 4,
    'AC_Usage_Hours': 8.0,
    'Fan_Usage_Hours': 12.0,
    'Refrigerator_Usage_Hours': 24.0,
    'Washing_Machine_Usage_Hours': 1.0,
    'Water_Motor_Usage_Hours': 0.5,
    'Lighting_Hours': 8.0,
    'Daily_Appliance_Count': 10,
    'Machinery_Load_kW': 0,
    'Machinery_Usage_Hours': 0,
    'Outdoor_Temperature_C': 35.0,
    'Electricity_Unit_Price': 35,
    'Day_of_Week': 5,
    'Season': 2,
    'Is_Holiday': 0,
    'Total_Appliance_Hours': 53.5,
    'Appliance_Hours_per_Person': 8.92,
    'Rooms_per_Person': 0.67,
    'High_AC_Usage': 0,
    'High_Temperature': 1,
    'Low_Temperature': 0,
    'Appliance_Count_per_Person': 1.67,
    'Is_Weekend': 0,
    'Efficiency_Score': 1.73,
    'AC_Fan_Ratio': 0.62
}

prediction = predict_consumption(example_input)

print(f"\nPredicted Daily Consumption: {prediction:.2f} kWh")
print(f"Estimated Monthly Consumption: {prediction * 30:.2f} kWh")
print(f"Estimated Monthly Cost (at 35 Rs/kWh): Rs. {prediction * 30 * 35:.2f}")

print("\n" + "="*80)
print("ENERGY OPTIMIZATION RECOMMENDATIONS")
print("="*80)

recommendations = generate_recommendations(example_input, prediction, 35)
for i, rec in enumerate(recommendations, 1):
    print(f"\n{i}. [{rec['category']}]")
    print(f"   Recommendation: {rec['recommendation']}")
    print(f"   Potential Savings: {rec['potential_savings']}")

# STEP 18: REPORT GENERATION

print("\n" + "="*80)
print("REPORT GENERATION")
print("="*80)

def generate_report():
    report_lines = []
    report_lines.append("="*80)
    report_lines.append("SMART ELECTRICITY CONSUMPTION PREDICTION SYSTEM")
    report_lines.append("Task ML-3 - Teyzix Core Internship (June Batch)")
    report_lines.append("="*80)
    report_lines.append("")

    report_lines.append("1. DATASET SUMMARY")
    report_lines.append("-"*50)
    report_lines.append(f"Total Records: {len(df_cleaned)}")
    report_lines.append(f"Total Features: {len(df_cleaned.columns)}")
    report_lines.append(f"Target Variable: Daily_Electricity_Consumption_kWh")
    report_lines.append("")
    report_lines.append("Target Variable Statistics:")
    report_lines.append(f"  Mean: {df_cleaned['Daily_Electricity_Consumption_kWh'].mean():.2f} kWh")
    report_lines.append(f"  Median: {df_cleaned['Daily_Electricity_Consumption_kWh'].median():.2f} kWh")
    report_lines.append(f"  Std Dev: {df_cleaned['Daily_Electricity_Consumption_kWh'].std():.2f} kWh")
    report_lines.append(f"  Min: {df_cleaned['Daily_Electricity_Consumption_kWh'].min():.2f} kWh")
    report_lines.append(f"  Max: {df_cleaned['Daily_Electricity_Consumption_kWh'].max():.2f} kWh")
    report_lines.append("")

    report_lines.append("2. MODEL PERFORMANCE")
    report_lines.append("-"*50)
    report_lines.append(f"Best Model: {best_model_name}")
    report_lines.append("")
    report_lines.append("All Models Performance:")
    for model in results_df.index:
        report_lines.append(f"  {model}:")
        report_lines.append(f"    MAE: {results_df.loc[model, 'MAE']:.4f}")
        report_lines.append(f"    RMSE: {results_df.loc[model, 'RMSE']:.4f}")
        report_lines.append(f"    R2 Score: {results_df.loc[model, 'R2']:.4f}")
        report_lines.append(f"    Time: {results_df.loc[model, 'Time']:.2f}s")
    report_lines.append("")

    report_lines.append("3. CROSS-VALIDATION RESULTS")
    report_lines.append("-"*50)
    report_lines.append(f"{best_model_name}:")
    report_lines.append(f"  Mean R2 Score: {cv_scores.mean():.4f}")
    report_lines.append(f"  Std R2 Score: {cv_scores.std():.4f}")
    report_lines.append("")

    if hasattr(best_model, 'feature_importances_'):
        report_lines.append("4. FEATURE IMPORTANCE (Top 10)")
        report_lines.append("-"*50)
        for idx, row in feature_importance_df.head(10).iterrows():
            report_lines.append(f"  {row['Feature']:30} | {row['Importance']:.4f}")
        report_lines.append("")

    report_lines.append("5. ENERGY OPTIMIZATION RECOMMENDATIONS")
    report_lines.append("-"*50)
    report_lines.append("1. Set AC temperature to 26°C or higher")
    report_lines.append("2. Use energy-efficient appliances (5-star rated)")
    report_lines.append("3. Schedule heavy loads during off-peak hours (12 AM - 6 AM)")
    report_lines.append("4. Replace traditional bulbs with LED lighting")
    report_lines.append("5. Use natural ventilation when outdoor temperature is comfortable")
    report_lines.append("6. Regularly maintain appliances for optimal efficiency")
    report_lines.append("7. Monitor daily consumption to identify unusual patterns")
    report_lines.append("8. Use smart power strips to eliminate phantom loads")
    report_lines.append("9. Insulate your home to reduce AC usage")
    report_lines.append("10. Install solar panels for long-term savings")
    report_lines.append("")

    report_lines.append("="*80)
    report_lines.append("END OF REPORT")
    report_lines.append("="*80)

    return "\n".join(report_lines)

report_text = generate_report()
print(report_text)

with open('Project_Report.txt', 'w') as f:
    f.write(report_text)
print("\nReport saved to 'Project_Report.txt'")

# STEP 19: INTERACTIVE PREDICTION

print("\n" + "="*80)
print("INTERACTIVE PREDICTION MODE")
print("="*80)

def interactive_prediction():
    print("\nEnter the following details for prediction:")
    print("-"*50)

    try:
        inputs = {}

        print("\nHouse Type Options:")
        print("  0: Independent House, 1: Apartment, 2: Villa")
        print("  3: Outlet, 4: General Store, 5: Industry, 6: Factory")
        inputs['House_Type'] = int(input("Enter House Type (0-6): "))

        inputs['Family_Members_or_Staff'] = float(input("Number of Family Members/Staff: "))
        inputs['Number_of_Rooms_or_Units'] = float(input("Number of Rooms/Units: "))

        print("\n--- Appliance Usage (hours per day) ---")
        inputs['AC_Usage_Hours'] = float(input("AC Usage Hours (0-24): "))
        inputs['Fan_Usage_Hours'] = float(input("Fan Usage Hours (0-24): "))
        inputs['Refrigerator_Usage_Hours'] = float(input("Refrigerator Usage Hours (0-24): "))
        inputs['Washing_Machine_Usage_Hours'] = float(input("Washing Machine Usage Hours (0-24): "))
        inputs['Water_Motor_Usage_Hours'] = float(input("Water Motor Usage Hours (0-24): "))
        inputs['Lighting_Hours'] = float(input("Lighting Hours (0-24): "))
        inputs['Daily_Appliance_Count'] = float(input("Number of Appliances Used Daily: "))

        print("\n--- Industrial Usage (if applicable) ---")
        inputs['Machinery_Load_kW'] = float(input("Machinery Load (kW) [0 if none]: "))
        inputs['Machinery_Usage_Hours'] = float(input("Machinery Usage Hours [0 if none]: "))

        inputs['Outdoor_Temperature_C'] = float(input("\nOutdoor Temperature (°C): "))

        print("\nDay of Week Options:")
        print("  0: Monday, 1: Tuesday, 2: Wednesday")
        print("  3: Thursday, 4: Friday, 5: Saturday, 6: Sunday")
        inputs['Day_of_Week'] = int(input("Enter Day of Week (0-6): "))

        print("\nSeason Options:")
        print("  0: Winter, 1: Spring, 2: Summer, 3: Monsoon")
        inputs['Season'] = int(input("Enter Season (0-3): "))

        inputs['Is_Holiday'] = int(input("Is Holiday? (0: No, 1: Yes): "))
        inputs['Electricity_Unit_Price'] = float(input("Electricity Unit Price (Rs/kWh): "))

        total_hours = (inputs['AC_Usage_Hours'] + inputs['Fan_Usage_Hours'] +
                       inputs['Refrigerator_Usage_Hours'] + inputs['Washing_Machine_Usage_Hours'] +
                       inputs['Water_Motor_Usage_Hours'] + inputs['Lighting_Hours'])

        inputs['Total_Appliance_Hours'] = total_hours
        inputs['Appliance_Hours_per_Person'] = total_hours / (inputs['Family_Members_or_Staff'] + 1)
        inputs['Rooms_per_Person'] = inputs['Number_of_Rooms_or_Units'] / (inputs['Family_Members_or_Staff'] + 1)
        inputs['High_AC_Usage'] = 1 if inputs['AC_Usage_Hours'] > 8 else 0
        inputs['High_Temperature'] = 1 if inputs['Outdoor_Temperature_C'] > 35 else 0
        inputs['Low_Temperature'] = 1 if inputs['Outdoor_Temperature_C'] < 15 else 0
        inputs['Appliance_Count_per_Person'] = inputs['Daily_Appliance_Count'] / (inputs['Family_Members_or_Staff'] + 1)
        inputs['Is_Weekend'] = 1 if inputs['Day_of_Week'] in [5, 6] else 0
        inputs['Efficiency_Score'] = total_hours / 25
        inputs['AC_Fan_Ratio'] = inputs['AC_Usage_Hours'] / (inputs['Fan_Usage_Hours'] + 1)

        prediction = predict_consumption(inputs)

        print("\n" + "="*80)
        print("PREDICTION RESULTS")
        print("="*80)
        print(f"Predicted Daily Consumption: {prediction:.2f} kWh")
        print(f"Predicted Monthly Consumption: {prediction * 30:.2f} kWh")
        print(f"Predicted Monthly Cost: Rs. {prediction * 30 * inputs['Electricity_Unit_Price']:,.2f}")

        print("\n" + "="*80)
        print("ENERGY OPTIMIZATION RECOMMENDATIONS")
        print("="*80)
        recommendations = generate_recommendations(inputs, prediction, inputs['Electricity_Unit_Price'])
        for i, rec in enumerate(recommendations, 1):
            print(f"\n{i}. [{rec['category']}]")
            print(f"   {rec['recommendation']}")
            print(f"   Potential Savings: {rec['potential_savings']}")

        return prediction

    except ValueError as e:
        print(f"Error: Invalid input. Please enter numeric values. Error: {e}")
        return None
    except Exception as e:
        print(f"Error: {e}")
        return None

print("\nWould you like to run interactive prediction?")
choice = input("Enter 'y' for yes, anything else to skip: ")
if choice.lower() == 'y':
    interactive_prediction()
else:
    print("Skipping interactive prediction.")

# STEP 20: FINAL SUMMARY

print("\n" + "="*80)
print("PROJECT COMPLETED SUCCESSFULLY!")
print("="*80)
print("")
print("="*40 + " SUMMARY " + "="*40)
print(f"Dataset: {len(df_cleaned)} records, {len(df_cleaned.columns)} features")
print(f"Features Created: {len(new_features)} new features")
print(f"Models Tested: {len(models)} models")
print(f"Best Model: {best_model_name}")
print(f"  R2 Score: {results_df.loc[best_model_name, 'R2']:.4f}")
print(f"  RMSE: {results_df.loc[best_model_name, 'RMSE']:.4f} kWh")
print(f"  MAE: {results_df.loc[best_model_name, 'MAE']:.4f} kWh")
print(f"Cross-Validation R2: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
print("="*80)
print("")
print("Files Created:")
print("  - models/electricity_consumption_model.pkl")
print("  - models/scaler.pkl")
print("  - models/feature_columns.pkl")
print("  - models/label_encoders.pkl")
print("  - Project_Report.txt")
print("  - smart_electricity_consumption_cleaned.csv")
print("  - smart_electricity_consumption_featured.csv")
print("")
print("="*80)
print("THANK YOU FOR COMPLETING THE PROJECT!")
print("="*80)

print("\n" + "="*80)
print("DOWNLOAD FILES")
print("="*80)
print("To download files, click on the 'Files' icon on the left sidebar")
print("Right-click on any file and select 'Download'")
print("\nImportant files to download:")
print("  - Project_Report.txt")
print("  - smart_electricity_consumption_cleaned.csv")
print("  - smart_electricity_consumption_featured.csv")
print("  - models/ (entire folder)")
```
